Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection and Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data into a pandas dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
# print the first 5 dataframe
data_frame.head()
```

```python
# Adding the target to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
# Finding the number of rows and columns of the dataframe
data_frame.shape
```

```python
# Getting some information about the dataframe
data_frame.info()
```

```python
# Inspecting the dataframe contains any missing values
data_frame.isnull().sum()
```

```python
# Statistical measures of the dataframe
data_frame.describe()
```

```python
# Finding the number of value of each type
data_frame['label'].value_counts()
```

1 -> Benign

0 -> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and target

```python
X = data_frame.drop(columns='label',axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data a testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

#**Building the Nural Network**

Tensor Flow is a deep learning library developed by google. We can create Neural Networks easily and there are several library like pytorch, microsoft cognition tookit.

Keras is not a standalone library and its the rapper of the tensor flow library.

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3) # Our accuracy value may change each time we run a neural network
# so to avoid that we set a seed to keep the accuracy value each time we run the neural network
from tensorflow import keras
```

```python
# Setting up the layers of the Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), # we are creating this flatten array in order to convert the data into single dimentional array.
    keras.layers.Dense(20,activation='relu'), # we are creating an hidden layer with 20 neuron.
    keras.layers.Dense(2,activation='sigmoid') # we are creating an output layer with 2 neuron as we have 2 classes.
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

Visualizing the accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')

plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')

plt.legend(['training data','validation data'],loc='upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std,Y_test)
print(accuracy)
```

model.predict() gives the prediction probability of each class for that data point.

```python
# argmax function
# it basically gives the maximum value from a list.
my_list = [0.56,0.25]

index_of_max_value = np.argmax(my_list)
print(my_list)
print(index_of_max_value)
```

```python
# converting the prediction probability to class labels
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

**Building the predictive system **

```python
input_data = (17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshaping the array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0]==0):
  print("The Tumor is Malignant")
else:
  print("The Tumor is Benign")
```
