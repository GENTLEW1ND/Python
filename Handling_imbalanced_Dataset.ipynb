**Imbalanced Dataset:**

  A dataset with an unequal class distribution.

```python
# Importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to pandas dataFrame
credit_card_data = pd.read_csv("/content/credit_data.csv")
```

```python
credit_card_data.head()
```

```python
# distribution the two classes
credit_card_data['Class'].value_counts()
```

This is Highly imblanced dataset

as the number of 0s is more than the number of 1s.

0 represents legit trasactions

1 represents fraudulent transaction

```python
# separating the ligit and fraudulent transaction
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

To handle imblanced transaction.

Build a sample dataset containing similar distribution of legit and fraudulent transaction.

Number of fraudulent transaction transaction is 213

```python
legit_sample = legit.sample(n=213)
```

```python
legit_sample.shape
```

Concatinate the two dataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
