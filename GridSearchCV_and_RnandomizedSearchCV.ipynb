```python
#importing the dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

We will be working on the breast cancer dataset

```python
# Loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# printing the first 5 rows of the dataframe
data_frame.head()
```

```python
# Adding the traget column in the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# Checking if there is any missing values
data_frame.isnull().sum()
```

```python
# Checking the distribution of the target variable
data_frame['label'].value_counts()
```

1 -> Benign

0 -> Malignant

Separating the features and target

```python
X = data_frame.drop(columns='label',axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

#**GridSearchCV**

GridSearchCV is used for determining the best parameters for our model.

```python
# Loading the SVC model
model = SVC()
```

```python
# Hyperparameters
parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# grid search
classifier = GridSearchCV(model,parameters,cv=5)
```

```python
# Fitting the data to our model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# best parameter
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# Determining the highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# Loading the results to a pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

Higest Accuracy = 95.2%

Best Parameter = {'C':10,'Kernel':'linear'}

# **RandomizedSearchCV**

```python
# Loading the SVC model
model = SVC()
```

```python
# Hyperparameters
parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# randomized search
classifier = RandomizedSearchCV(model,parameters,cv=5)
```

```python
# Fitting the data to our model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# best parameter
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# Determining the highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# Loading the results to a pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
randomized_search_result
```

Higest Accuracy = 95.0%

Best Parameter = {'C':5,'Kernel':'linear'}
